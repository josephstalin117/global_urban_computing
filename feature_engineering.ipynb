# 样例：某一个站点的处理

```python
# Data Loading
import os
import pandas as pd
import numpy as np

metro_train_file_path = '/home/zhibo/data/hexi/hexi2/global_urban_computing/data/temp'
metro_train_data = pd.read_csv(metro_train_file_path+'/station_46.csv') 

metro_train_data['timestamp'] = pd.to_datetime(metro_train_data['timestamp'])
metro_train_data['week_day'] = metro_train_data['timestamp'].dt.dayofweek
metro_train_data['current_hour'] = metro_train_data['timestamp'].dt.hour
metro_train_data[['lag_yesterday_in','lag_yesterday_out']] = metro_train_data[['inNums','outNums']].shift(periods=144).fillna(-1).astype(int)
metro_train_data[['lag_last_week_in','lag_last_week_out']] = metro_train_data[['inNums','outNums']].shift(periods=144*7).fillna(-1).astype(int)
metro_train_data['is_holiday'] = metro_train_data['week_day'].isin([5,6])
metro_train_data.set_index('timestamp',inplace=True)
metro_train_data.loc['2019-01-01','is_holiday']=True
metro_train_data['is_holiday'] = metro_train_data['is_holiday']*1
metro_train_data.head()
```

# 主要代码

```python
import os
import pandas as pd
metro_train_file_path = '/home/zhibo/data/hexi/hexi2/global_urban_computing/data'

for i in range(81):
    metro_train_data = pd.read_csv(metro_train_file_path + '/temp/station_' + str(i) + '.csv') 
    metro_train_data['timestamp'] = pd.to_datetime(metro_train_data['timestamp'])
    metro_train_data['week_day'] = metro_train_data['timestamp'].dt.dayofweek
    metro_train_data['current_hour'] = metro_train_data['timestamp'].dt.hour
    metro_train_data[['lag_yesterday_in','lag_yesterday_out']] = metro_train_data[['inNums','outNums']].shift(periods=144).fillna(-1).astype(int)
    metro_train_data[['lag_last_week_in','lag_last_week_out']] = metro_train_data[['inNums','outNums']].shift(periods=144*7).fillna(-1).astype(int)
    metro_train_data['is_holiday'] = metro_train_data['week_day'].isin([5,6])
    metro_train_data.set_index('timestamp',inplace=True)
    metro_train_data.loc['2019-01-01','is_holiday']=True
    metro_train_data['is_holiday'] = metro_train_data['is_holiday']*1
    metro_train_data.to_csv(metro_train_file_path + '/data_feature/st_' + str(i) + '.csv')
```
