```python
#Hourly
hourly = metro_train_data.resample('H').mean()

#Daily
daily = metro_train_data.resample('D').mean()

#Weekly
weekly = metro_train_data.resample('W').mean()

fig,axs = plt.subplots(3,1)

hourly[['inNums','outNums']].plot(figsize = (15,8), title = "Hourly", fontsize = 14, ax = axs[0])
daily[['inNums','outNums']].plot(figsize = (15,8), title = "Daily", fontsize = 14, ax = axs[1])
weekly[['inNums','outNums']].plot(figsize = (15,8), title = "Weekly", fontsize = 14, ax = axs[2])
```
